```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# 1) How many neighborhood groups are available and which shows up the most?
 # air_bnb.groupby('neighbourhood_group').count().sort_values(['id'], ascending=False)[['id']] 
    # -> boylede tum mahalleleri bir col'da countlari bir col'da veriyor

air_bnb[['neighbourhood_group']].describe()
##
   # There are 5 neighbourdhood groups and Manhattan is the one shows up the most.
```

```python
# 2) Are private rooms the most popular in manhattan?
air_bnb[['neighbourhood_group', 'room_type']].describe()

## 
  # No, entire home/apts are popular than privates rooms in Manhattan. 
```

```python
manh_bool = air_bnb['neighbourhood_group'] == 'Manhattan' # bu eger mallalle manhattansa true veriyor, degilse false
only_manh = air_bnb[manh_bool] # bu sadece mahallenin manhattan oldugu datalari full tablo olarak gosterdi
only_manh['room_type'].describe()
```

```python
# 3) Which hosts are the busiest and based on their reviews?
busiest_host = air_bnb.groupby(['host_id','host_name', 'neighbourhood_group']).sum()
busiest_host.sort_values('number_of_reviews', ascending=False)[['number_of_reviews']].head(5)

## 
  # Maya with id number 37312959 is the busisest hosts based on the review numbers ->2273
```

```python
maya = air_bnb[air_bnb['host_id'] == 37312959]
maya.groupby(['host_name', 'neighbourhood_group', 'last_review', 'reviews_per_month']).sum().sort_values('number_of_reviews', ascending=False)[['number_of_reviews']].head(10)
```

```python
maya.groupby(['host_name'])[['price']].sum().sort_values(['price'], ascending=False)
```

```python
# 4) Which neighorhood group has the highest average price?

    # avg = air_bnb.groupby('neighbourhood_group').mean() # bu butun mahallelerdeki tum col'larin ortalamalarini veriyor
    # avg_price = avg[['price']] # -> bu sadece mahalle adi ve ortamala fiyat kolonlarini veriyor belli bir sira olmadan
    # avg_siralama = avg_price.sort_values(['price'], ascending=False)
    
air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values(['price'], ascending=False)

##
  # Manhattan with an average price of $196.88
```

```python
avg_price = air_bnb.groupby('neighbourhood_group')

brooklyn = avg_price.get_group('Brooklyn')['price'].mean()
manhattan = avg_price.get_group('Manhattan')['price'].mean()
staten_island = avg_price.get_group('Staten Island')['price'].mean()
queens = avg_price.get_group('Queens')['price'].mean()
bronx = avg_price.get_group('Bronx')['price'].mean()

plt.bar(
    air_bnb['neighbourhood_group'].unique(),
    [brooklyn, manhattan, queens, staten_island, bronx],
    color='b'
)
plt.xlabel('Neighbourhood Groups')
plt.ylabel('Average Price')
```

```python
# Which neighborhood group has the highest total price?
highest_nhood = air_bnb.groupby('neighbourhood_group').sum()[['price']].sort_values(['price'], ascending=False)
highest_nhood
```

```python
highest_nhood.plot(kind="barh", fontsize=8)

plt.xlabel('Average Price')
plt.ylabel('Total Price')
```

```python
# Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_name', 'host_id'])[['price']].sum().sort_values(['price'], ascending=False).head()
```
